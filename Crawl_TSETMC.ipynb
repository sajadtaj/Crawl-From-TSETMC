# Get Data From TSETMC by requests

### import package

```python
import requests
import pandas as pd
```

## Share Holders

```python
stock = 'وبملت'
code = 778253364357513
date = 20230614
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36 Edg/95.0.1020.53'}
url = 'http://cdn.tsetmc.com/api/Shareholder/{}/{}'.format(code, date)

response        = requests.get(url, headers=headers)
shareholders    = response.json()['shareShareholder']
shareholders_df = pd.DataFrame(shareholders)[::-1]
grouped         = shareholders_df.groupby('shareHolderID')

shareholders_df['Change_Number']  = grouped['numberOfShares'].diff()
shareholders_df['Change_Percent'] = grouped['perOfShares'].diff()
shareholders_df                   = shareholders_df[shareholders_df['dEven'] == 20230615 ]
shareholders_df                   = shareholders_df.drop(['change', 'changeAmount'], axis=1 )
```

```python
shareholders_df
```

## Trade History

```python
stock = 'وبملت'
code = 13611044044646901
date = 20230913
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36 Edg/95.0.1020.53'}
url = 'http://cdn.tsetmc.com/api/Trade/GetTradeHistory/{}/{}/true'.format(code, date)
response = requests.get(url, headers=headers)
tradeHistory = response.json()['tradeHistory']
tradeHistory_df = pd.DataFrame(tradeHistory)[::-1]
```
